```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
df = pd.read_csv("CC for Attributes P3.csv",header=0)
df.head()
```

```python
c = df.number_of_nonconformities
c_bar = c.mean()
print(c_bar)
```

```python
UCL_c = c_bar + 3*math.sqrt(c_bar)
CL_c = c_bar
LCL_c = max(0,c_bar - 3*math.sqrt(c_bar))
print("UCL_c: ",UCL_c,'\n',"CL_c: ",CL_c,'\n',"LCL_c: ",LCL_c)
```

```python
if LCL_c < 0:
    LCL_c = 0
print("UCL_c: ",UCL_c,'\n',"CL_c: ",CL_c,'\n',"LCL_c: ",LCL_c)
```

```python
ax = plt.figure().gca()
ax.plot(c.index,c,marker="o")
ax.axhline(y = UCL_c, color = 'r', linestyle = '-')
ax.axhline(y = LCL_c, color = 'r', linestyle = '-')
# add text showing the value of UCL and LCL
ax.text(0.05, 0.85, 'UCL = {}\nLCL = {}\nCL = {}'.format(UCL_c, LCL_c, CL_c),
         verticalalignment='top', horizontalalignment='left',
         transform=ax.transAxes, color='red', fontsize=10)

ax.axhline(y = CL_c, color = 'r', linestyle = '-')
from matplotlib.ticker import MaxNLocator
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(True)
plt.title('C chart, Phase 1')
plt.xlabel('Sample number, $i$')
plt.ylabel('Number of Nonconformities')

plt.show()
```

```python
Mask_to_remove_1=(c.ge(UCL_c) | c.le(LCL_c))
```

```python
c.iloc[Mask_to_remove_1] = np.nan
c_bar = c.mean()
UCL_c = c_bar + 3*math.sqrt(c_bar)
CL_c = c_bar
LCL_c =max(0,c_bar - 3*math.sqrt(c_bar)) 
print("UCL_c: ",UCL_c,'\n',"CL_c: ",CL_c,'\n',"LCL_c: ",LCL_c)
```

```python
if LCL_c < 0:
    LCL_c = 0
print("UCL_c: ",UCL_c,'\n',"CL_c: ",CL_c,'\n',"LCL_c: ",LCL_c)
```

```python
ax = plt.figure().gca()
ax.plot(c.index,c,marker="o")
ax.axhline(y = UCL_c, color = 'r', linestyle = '-')
ax.axhline(y = LCL_c, color = 'r', linestyle = '-')
# add text showing the value of UCL and LCL
ax.text(0.05, 0.85, f'UCL = {UCL_c}\nLCL = {LCL_c}\nCL = {CL_c}',
         verticalalignment='top', horizontalalignment='left',
         transform=ax.transAxes, color='red', fontsize=10)

ax.axhline(y = CL_c, color = 'r', linestyle = '-')
from matplotlib.ticker import MaxNLocator
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(True)
plt.title('C Chart, Phase 1')
plt.xlabel('Sample number, $i$')
plt.ylabel('Number of Nonconformities')

plt.show()
```
