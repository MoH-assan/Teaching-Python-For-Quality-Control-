```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(12345)
from scipy import stats
```

```python
def AppendixVI(n,factor=['A','n','A','A2','A3','c4','1/c4','B3','B4','B5','B6','d2','1/d2','d3','D1','D2','D3','D4'][-4]):
    #Note that appendix AppendixVI is based on 0.997 confidence level (i.e. z_zalpha_2`~2.97)'
    '''
    Just an alternative for looking up the table
    n: sample size
    factor: factor to be checked at n 
    '''
    AppendixVI=pd.read_csv('AppendixVI.csv',index_col='n')
    return AppendixVI[factor][n]
```

```python
def closest_lower_index(lst, target):
    lower_values = [value for value in lst if value < target]
    if not lower_values:
        return None
    return lst.index(max(lower_values))
```

# Q1_A

```python
n=150 #Here we are dealing with constant sample size #Given
P_bar=0.0185 #Average proportion of defective items #Given
alpha=0.002 #Given
```

```python
mode='two-sided' #or 'one-sided'
def LCL_np_chart(n,P_bar,alpha,printing=True):
    for Possible_LCL in np.arange(np.floor(n*P_bar),-1,-1):
        if printing: print(f'Testing LCL of {Possible_LCL} ......')
        p_LCL=stats.binom.cdf(k=Possible_LCL,n=n,p=P_bar) #probability of getting more less defectives than current possible LCL defects
        type_1_error=p_LCL*2
        if printing: print(f'Resultant type_1_error is {type_1_error}')
        if type_1_error<=alpha:
            if printing: print('**************Found LCL**************')
            if printing: print(f'At LCL {Possible_LCL}, type_1_error is {type_1_error} which is less than alpha of {alpha}')
            return Possible_LCL,'two-sided'
        elif Possible_LCL==0:
            if printing: print('************** No Found LCL**************')
            if printing: print(f'Having Zero Defectives is Likely For Your Case, Switch to "one-sided" Test')
            if printing: print(f'Setting LCL to 0')
            return 0,'one-sided'
            break
```

```python
def UCL_np_chart(n,P_bar,alpha,mode='two-sided',printing=True):
    if mode not in ['two-sided','one-sided']:

        if printing: print('Mode should be either two-sided or one-sided')
        return None
    else:
        if printing: print(f'**************{mode} is Activated**************')
    for Possible_UCL in np.arange(np.ceil(n*P_bar),n+1):
        if printing: print(f'Testing UCL of {Possible_UCL} ......')
        p_ucl=1-stats.binom.cdf(k=Possible_UCL,n=n,p=P_bar) #probability of getting more defectives UCL defects
        if mode=='two-sided':
            type_1_error=p_ucl*2
        elif mode=='one-sided':
            type_1_error=p_ucl
        if printing: print(f'Resultant type_1_error is {type_1_error}')
        if type_1_error<=alpha:
            if printing: print('**************Found UCL**************')
            if printing: print(f'At UCL {Possible_UCL}, type_1_error is {type_1_error} which is less than alpha of {alpha}')
            return Possible_UCL
            break
```

```python
LCL,mode_recommended=LCL_np_chart(n=150,P_bar=0.0185,alpha=0.002)
```

```python
UCL=UCL_np_chart(150,0.0185,0.002,mode=mode_recommended,printing=True)
```
