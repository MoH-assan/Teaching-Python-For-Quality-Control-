```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(12345)  
from scipy import stats
import imageio
from PIL import Image
```

Given 

```python
total_defectives=1200
m=30
n=400
```

Q2)a

```python
P_bar=total_defectives/(m*n)
print(f'P_bar is equal to {P_bar}')
```

Since no alpha is give so we can assume we can used the 3 sigma rule to construct the np-chart

```python
UCL=n*P_bar+3*np.sqrt(n*P_bar*(1-P_bar))
LCL=max(0,n*P_bar-3*np.sqrt(n*P_bar*(1-P_bar)))
CL=P_bar*n
print(f'UCL is equal to {UCL}')
print(f'LCL is equal to {LCL}')
print(f'CL is equal to {CL}')
```

Q2)b

```python
p_1=0.15
```

```python
beta=stats.binom.cdf(UCL-1,n=n,p=p_1)-stats.binom.cdf(LCL,n=n,p=p_1) #Note that the UCL-1
print(f'beta is equal to {beta}')
power=1-beta
print(f'power is equal to {power}')
```

Q2)c

```python
ARL_1=1/(power)
print(f'ARL_1 is equal to {ARL_1}')
```
