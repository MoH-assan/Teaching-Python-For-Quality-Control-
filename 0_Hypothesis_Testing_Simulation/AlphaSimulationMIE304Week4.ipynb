```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#Define_Null  (aka population)
z_o=0
sigma_o=1
n=10 # Sample size, could be anything 

signa_means=sigma_o/np.sqrt(n) # std of "Samples means distribution" 
UCL=1.976*signa_means
LCL=-1.976*signa_means

```

```python
#Sample a single sample
sample=norm.rvs(loc=z_o,scale=sigma_o,size=n)
sample_mean=np.mean(sample)
print(sample_mean)
```

```python
m=20 #Number of Samples
list_samples_means=[]
for counter in range(0,m,1):  #range(m)
    sample=norm.rvs(loc=z_o,scale=sigma_o,size=n)
    sample_mean=np.mean(sample)
    #print(sample_mean)
    list_samples_means.append(sample_mean)
```

```python
Nr=np.sum(abs(np.array(list_samples_means))>UCL)
alpha=Nr/m
print(f'Alpha is equal to {alpha}')
print('Alpha is equal to {alpha}')
```

```python
plt.hist(list_samples_means)
```

```python
list_m=[]
list_alpha=[]
for m in range(0,1000):
    #m=20 #Number of Samples
    list_samples_means=[]
    for counter in range(0,m,1):  #range(m)
        sample=norm.rvs(loc=z_o,scale=sigma_o,size=n)
        sample_mean=np.mean(sample)
        #print(sample_mean)
        list_samples_means.append(sample_mean)
    Nr=np.sum(abs(np.array(list_samples_means))>UCL)
    alpha=Nr/m
    list_m.append(m)
    list_alpha.append(alpha)
plt.plot(list_m,list_alpha)
```
