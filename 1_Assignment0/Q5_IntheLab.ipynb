```python
from  scipy.stats import chi2
import numpy as np
```

# Q5a

```python
x=[4532, 4606, 3511, 4201, 3392, 4639,
4021, 4722, 3470, 3100, 4212, 4165,]
n=len(x)
sigma_o=160000**0.5
alpha=0.1
sigma_samples=np.std(x,ddof=1) #Not given
print(sigma_samples)
```

```python
x_alpha_2=chi2.ppf(alpha/2,n-1)
print(x_alpha_2)
LCL=(n-1)*sigma_samples**2/x_alpha_2
print(LCL)
sigma_o**2>LCL
```

```python
x_1_alpha_2=chi2.ppf(1-alpha/2,n-1)
print(x_1_alpha_2)
UCL=(n-1)*sigma_samples**2/x_1_alpha_2
print(UCL)
sigma_o**2<UCL
```

# p-value

```python
x_observed=(n-1)*sigma_samples**2/sigma_o**2
print(x_observed)
```

```python
second_term=chi2.cdf(x_observed,df=n-1)
#print(second_term)

first_term=1-second_term
#print(first_term)

p_value=2*np.min([first_term,second_term])
print(p_value)
```
