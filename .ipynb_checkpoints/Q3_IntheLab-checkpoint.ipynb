```python
from  scipy.stats import norm
import numpy as np
```

# Q1A

```python
x=[41.7, 40.6, 41.0, 42.5, 40.4,
42.4, 42.7, 41.2, 42.6, 41.5,]
n=len(x)
mu_o=41
sigma_o=np.sqrt(0.64) #Given
```

```python
n
```

```python
x_bar=np.mean(x)
```

```python
z_obs=(x_bar-mu_o)/(sigma_o/np.sqrt(n))
```

```python
p_value=(1-norm.cdf(z_obs))*2
p_value
```

```python
Second_phi=norm.cdf(-z_alpha_2-delta*np.sqrt(n)/sigma_o)
print(Second_phi)

print(-z_alpha_2-delta*np.sqrt(n)/sigma_o)
```
